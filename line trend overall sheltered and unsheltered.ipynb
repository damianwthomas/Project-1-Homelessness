```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = 'HUD Subtotal Summary.csv'
hud_data = pd.read_csv(file)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.5,0.5,0.9,0.9])
axes.plot(hud_data['Year'],hud_data['Sum of Overall Homeless'],marker = 'o')
axes.plot(hud_data['Year'],hud_data['Sum of Sheltered Total Homeless'], marker = 'o')
axes.plot(hud_data['Year'],hud_data['Sum of Unsheltered Homeless'], marker = 'o')
axes.set_xlabel ('Years')
axes.set_ylabel ('Homeless Count')
axes.set_title ('Estimated Homeless Individuals in U.S')
axes.legend(loc=(1.1,0.5))
plt.show
```
